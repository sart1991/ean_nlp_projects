```python
DF=pd.read_csv('../data/reviews_booking.csv', low_memory=False)
DF.head(2)
```

```python
DF['Calificación']=DF['Calificación'].str.replace(',','.').astype(float)
```

```python
DF_inicial=DF[DF['Calificación']<=7]

DF_inicial=DF_inicial[DF_inicial['País'].isin(['Colombia'])]['Cosas Negativas']
DF_inicial=DF_inicial[DF_inicial.fillna('').apply(len)>60].sample(7500)

DF_inicial
```

```python
DF_Final=pd.DataFrame()

DF_Final['Comentarios']=DF_inicial
DF_Final['Clas']='Negativos'
DF_Final
```

```python
DF_inicial2=DF[DF['Calificación']>7]

DF_inicial2=DF_inicial2[DF_inicial2['País'].isin(['Colombia'])]['Cosas Positivas']
DF_inicial2=DF_inicial2[DF_inicial2.fillna('').apply(len)>60].sample(7500)

DF_final2=pd.DataFrame()

DF_final2['Comentarios']=DF_inicial2
DF_final2['Clas']='Positivos'
DF_final2
```

```python
DF_Final_Total = pd.concat([DF_Final, DF_final2])

DF_Final_Total
```

```python
import re
def normalizar_texto(texto):
    texto = texto.lower()
    texto = re.sub(r'[^a-záéíóúüñ\s]', '', texto)
    texto = re.sub(r'\d+', '', texto)
    texto = re.sub(r'\s+', ' ', texto).strip()
    return texto

DF_Final_Total['Comentarios_normalizados'] = DF_Final_Total['Comentarios'].apply(normalizar_texto)
```

```python
DF_Final_Total
```

```python
DF_Final_Total['Tokens'] = DF_Final_Total['Comentarios_normalizados'].apply(lambda texto: texto.split())
DF_Final_Total
```

```python
import nltk
from nltk.corpus import stopwords

stopwords_esp = stopwords.words('spanish')
DF_Final_Total['Tokensstopw'] = DF_Final_Total['Tokens'].apply(lambda tokens: [token for token in tokens if token not in stopwords_esp])
DF_Final_Total
```

```python
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(DF_Final_Total['Comentarios_normalizados'])

# Convertir la matriz TF-IDF a DataFrame
tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=vectorizer.get_feature_names_out()
)
word_weights = tfidf_df.mean(axis=0).sort_values(ascending=False)
print("Palabras más importantes según TF-IDF:")
print(word_weights.head(10))
```
